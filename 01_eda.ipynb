```python

def merge_data():
    print("üöÄ STARTING CINE-MATCH DATA PIPELINE")
    print("=" * 60)

    # ---------------------------------------------------------
    # STEP 1: LOAD & FILTER TMDB (The Master List)
    # ---------------------------------------------------------
    print(f"1Ô∏è‚É£  Loading TMDb from: {TMDB_PATH}")
    if not os.path.exists(TMDB_PATH):
        print("‚ùå Error: TMDb file not found.")
        return

    # Load full dataset
    df_tmdb = pd.read_csv(TMDB_PATH)
    total_raw = len(df_tmdb)
    print(f"   - Raw TMDb count: {total_raw:,}")

    # --- FILTER A: EXISTENCE CHECK (Validated by your analysis) ---
    print("   - filtering: Removing movies with no IMDb ID...")
    
    # Drop rows where imdb_id is NaN (missing)
    df_tmdb.dropna(subset=['imdb_id'], inplace=True)
    
    # Ensure they are strings and not empty
    df_tmdb['imdb_id'] = df_tmdb['imdb_id'].astype(str).str.strip()
    df_tmdb = df_tmdb[df_tmdb['imdb_id'] != '']
    
    count_after_id = len(df_tmdb)
    print(f"   - Valid IMDb IDs: {count_after_id:,} (Dropped {total_raw - count_after_id:,} junk records)")

    # --- FILTER B: QUALITY CHECK ---
    print("   - filtering: Applying Quality Gates (Votes >= 50 OR New)...")
    
    # Clean numeric columns
    df_tmdb['vote_count'] = pd.to_numeric(df_tmdb['vote_count'], errors='coerce').fillna(0)
    df_tmdb['popularity'] = pd.to_numeric(df_tmdb['popularity'], errors='coerce').fillna(0)
    
    # Keep if: Popular/Voted OR New Release (2024+)
    is_popular = (df_tmdb['vote_count'] >= 50) | (df_tmdb['popularity'] >= 5.0)
    is_recent = (df_tmdb['release_date'] >= '2024-01-01')
    
    df_tmdb_filtered = df_tmdb[is_popular | is_recent].copy()
    
    print(f"   - High-Quality TMDb count: {len(df_tmdb_filtered):,} (Dropped another {count_after_id - len(df_tmdb_filtered):,} low-quality rows)")

    # ---------------------------------------------------------
    # STEP 2: LOAD & FILTER IMDB (The Adult Flag Source)
    # ---------------------------------------------------------
    print(f"\n2Ô∏è‚É£  Loading IMDb from: {IMDB_PATH}")
    
    # We only need 'tconst' (ID), 'isAdult', and 'titleType'
    cols_needed = ['tconst', 'isAdult', 'titleType']
    
    df_imdb = pd.read_csv(
        IMDB_PATH, 
        sep='\t', 
        usecols=cols_needed, 
        dtype={'tconst': str, 'isAdult': float}, 
        na_values='\\N',
        low_memory=False
    )
    
    print(f"   - Raw IMDb count: {len(df_imdb):,}")
    
    # --- FILTER C: TYPE CHECK ---
    # Keep ONLY 'movie' and 'tvMovie' (removes 9M+ episodes)
    valid_types = ['movie', 'tvMovie']
    df_imdb_filtered = df_imdb[df_imdb['titleType'].isin(valid_types)].copy()
    
    print(f"   - Kept only 'movie' & 'tvMovie': {len(df_imdb_filtered):,} rows")

    # ---------------------------------------------------------
    # STEP 3: MERGE
    # ---------------------------------------------------------
    print("\n3Ô∏è‚É£  Merging Datasets...")
    
    # Standardize ID column name in IMDb for the merge
    df_imdb_filtered['tconst'] = df_imdb_filtered['tconst'].astype(str).str.strip()
    
    # LEFT JOIN: Keep our curated TMDb list, attach IMDb info
    merged = pd.merge(
        df_tmdb_filtered,
        df_imdb_filtered[['tconst', 'isAdult']], 
        left_on='imdb_id',
        right_on='tconst',
        how='left'
    )
    
    # ---------------------------------------------------------
    # STEP 4: FINALIZE
    # ---------------------------------------------------------
    print("   - Finalizing 'adult' flag...")
    
    # If isAdult == 1 -> True. Else -> False
    merged['adult'] = (merged['isAdult'] == 1.0)
    
    # Drop temp columns
    merged.drop(columns=['tconst', 'isAdult'], inplace=True, errors='ignore')
    
    # Save
    os.makedirs(os.path.dirname(OUTPUT_PATH), exist_ok=True)
    merged.to_csv(OUTPUT_PATH, index=False)
    
    print("=" * 60)
    print(f"‚úÖ SUCCESS! Clean data saved to: {OUTPUT_PATH}")
    print(f"   Final Count: {len(merged):,} movies")
    print(f"   Adult Content: {merged['adult'].sum()} movies flagged.")
    print("=" * 60)

if __name__ == "__main__":
    merge_data()
```

```python
import chromadb
from chromadb.utils import embedding_functions
import pandas as pd

# --- CONFIGURATION ---
# MUST match what you used in build_vector.py
PERSIST_DIR = "./chroma_db"
COLLECTION_NAME = "cine_match_v1"

def test_search(query_text):
    print(f"\nüîé Querying for: '{query_text}'...")
    
    # 1. Connect to the existing DB
    client = chromadb.PersistentClient(path=PERSIST_DIR)
    
    # 2. Get the collection
    embedding_func = embedding_functions.SentenceTransformerEmbeddingFunction(
        model_name="all-MiniLM-L6-v2"
    )
    
    try:
        collection = client.get_collection(
            name=COLLECTION_NAME,
            embedding_function=embedding_func
        )
    except Exception as e:
        print(f"‚ùå Error: Could not find collection. Did the build script finish? \n{e}")
        return

    # 3. Search
    results = collection.query(
        query_texts=[query_text],
        n_results=3,
        # SAFE SEARCH: Filter out adult content by default
        where={"adult": False} 
    )

    # 4. Display Results
    # Chroma returns lists of lists (because you can query multiple texts at once)
    # We just want the first list [0]
    for i in range(len(results['ids'][0])):
        title = results['metadatas'][0][i]['title']
        genres = results['metadatas'][0][i]['genres_display']
        dist = results['distances'][0][i] # Lower distance = Better match
        
        print(f"   {i+1}. {title} | Genres: {genres} | Score: {dist:.4f}")

# --- RUN THE TEST ---
if __name__ == "__main__":
    # Test 1: Specific Title (Should be #1 match)
    test_search("Swiss Army Man")
    
    # Test 2: Semantic Concept (No title match)
    test_search("A movie about a robot that falls in love")
```
